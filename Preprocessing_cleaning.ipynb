```python
#Titanic Data Preprocessing
#Step 1:Import lib
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
#Step 2: Load the titanic datastet

titanicdf=pd.read_csv("titanic.csv")
#Replace with correct
titanicdf
```

```python
#Step3:Show basic info and check missing values
print(titanicdf.info())
print(titanicdf.isnull().sum())
```

```python
titanicdf['Age'].fillna(titanicdf['Age'].mean(),inplace=True)
```

```python
titanicdf
```

```python
#filling missing values in cabin
#titanicdf['Cabin'].fillna(titanicdf['Cabin'].mode(),inplace=True)
```

```python
titanicdf['Cabin'] = titanicdf['Cabin'].fillna('Unknown')
titanicdf['Cabin'] = titanicdf['Cabin'].apply(lambda x: x[0])
```

```python
titanicdf
```

```python
#Step 4:Missing Values

#Fill Age with median
titanicdf['Age'].fillna(titanicdf['Age'].median(),inplace=True)
```

```python
titanicdf
```

```python
#Step5 Drop Irrelevant or Redundant Columns
#drop passengerId,Name,Ticket(not useful for ML directly)
titanicdf.drop(['PassengerId','Name','Ticket'],axis=1,inplace=True)
```

```python
#Step6 Encode categorical variables

#Use LabelEncoder for Sex, Embarked, and Cabin(simplified)
le=LabelEncoder()

titanicdf['Sex']=le.fit_transform(titanicdf['Sex'])
#male=1,female=0

```

```python
titanicdf
```

```python
titanicdf['Embarked']=le.fit_transform(titanicdf['Embarked']) #C=0,Q=1,S=2

```

```python
titanicdf
```

```python
#For 'Cabin', simplify to just first letter(deck)
#titanicdf['Age']=titanicdf['Age'].apply(lambda x: x[0]) #Eg, C123..gives..C
```

```python
#For 'Cabin', simplify to just first letter(deck)
titanicdf['Cabin']=titanicdf['Cabin'].apply(lambda x: x[0]) #Eg, C123..gives..C
```

```python
titanicdf['Cabin']=le.fit_transform(titanicdf['Cabin'])
```

```python
titanicdf
```

```python
#Final Check
#Confirm all preprocessing

print(titanicdf.head())
print(titanicdf.info())
print(titanicdf.isnull().sum())
```
